# Minimal VARX: The $\AR$-state-space approach to time series analysis.
In this project we show implications an autoregressive state-space approach to time series and linear system identification.
The traditional state-space approach expresses $\by$ a function of inputs and innovations. This is the traditional approach in control theory, but in forecasting the link between past and future observation is not explicit, making it is difficult for both model fitting and forecasting.


We advocate the $\AR$ state-space approach. It is expressed as
$$\by = \bH(\bI-\bF L)^{-1}\bG L \by + \bep$$
By a theorem of Kalman, $\VAR, \VARMA$ and their exogenous counterparts are all expressible in this form. For the $\VARX(1)$ the minimal $\AR$ state-space approach is exactly the reduced rank regression, and the estimation procedure for reduced rank regression could be generalized. We show the (negative) likelhood function could be expressed as a determinant ratio. In the reduced-rank case maximizing this likelihood function leads to canonical- correlation-analysis. In the general case we must use a numerical procedure, but both the gradient and the Hessian could be computed explicitly for MLE.

For the $\VAR(p)$ case, $\bF$ could be reduced to a Jordan matrix with $\bF^p=0$. This allows an explicit classication of possible $minimal$ state-space. This helps providing a framework for parameter reductions going beyond reduced-rank regression. While in the $p=1$ case we factor the regression matrix $\bPhi$ to a product $\bH\bG$, in the general case we distribute the ranks $d_i$ on to each power $1\leq i\leq p$. These ranks define the complexity of the model. The most expensive model in term of parameter is one with all the rank concentrated at the $\Phi_p$ parameter, the least expensive is one where $\Phi_p$ of rank $1$ (and no lower power get allocate any rank). This provides an intuitive look at complexity of state-space model.

More interestingly, we can reduce the search space further by using matrices $\bS$ that commute with $\bF$. $\bG$ could be made to satisfy two sets of orthogonal relations. Hence we can apply manifold optimization technique for high dimensional problem. It turns out the manifold in question is an interesting mathematical object by itself, a vector bundle over a flag manifold. However, there is no need to get to manifold optimization to use our model.

Now we can import the modules and packages. 

We have a **polynomial_matrix** module built on the numpy polynomial package that provides basic functionality to multiply polynomial matrices, computing determinant and roots to check for stability. It can also do Smith normal form, with some caveats as decomposition could be sensitive to numerical tolerance.
We will not go in to the package in any detail, but sometime the lag polynomial $\Phi$ could be given as an object in the polynomial matrix class.

**utils** contains a number of utility functions. VAR_sim is to simulate an autoregressive model, where the noise covariance matrix is given by $\bO^T  \bD  \bO$ and autoregressive terms is given by Phi. There is a utility to generate a random orthogonal matrix. list_all_psi_hat and psi_hat_to_psi are functions to list all possible nilpotent Jordan matrix configurations.

**minimal_varx** is the main module. Its main object is varx_minimal_estimator. Given $\Psi$ and $m$, it constructs the basic structure, including aggregate rank ($\sum d_i$), McMillan dimension. It can compute $\kappa$. Set it up with some covariant matrices and $\bG$ it computes the optimal negative log likelihood for that $\bG$ via $calc\_states$. That function also compute gradient. Hessian product is computed via a call to $hessian\_product$.

It has a few fit functions, $simple\_fit$ does not use gradient, $gradient\_fit$ uses gradient, $hessian\_fit$ uses Hessian product and $manifold\_fit$ uses manifold optimization (we set up the later for the manifold case only, not the vector bundle case - and not using the full symmetry.)

Central to our set up is a Jordan matrix configuration $\Psi$. For most of the paper we are assuming that a Jordan matrix configuration is picked. We will show a few operations we can perform with Jordan matrix configuration:

```
Psi =  [(3, 2), (2, 1), (1, 2)]
m = 5
es = varx_minimal_estimator(Psi, m=m)
# Jordan matrix
print(pm.calc_Jordan_matrix(es.Psi, 0))
# McMilan degree
print('McMillan degree is %d' % es.mm_degree)
# aggregate rank sum d_i
print('Aggregate rank is %d' %es.agg_rnk)
```

In the above, $\Psi$ specifies a Jordan matrix of form $\bK(3,2)\oplus \bK(2, 1)\oplus \bK(1, 2)$. The total aggregate rank $\fll$ is $5$, same as $m$. So in this case there is no extra dimension. The McMillan degree which is the dimension of $\bF$ is $10$.

Another example is the LQ factorization applied to multi-lag system. Given a $\Psi$, and $\bG$ of size $\nmin\times m$ can be factorized to $\bS, \bW$ with $ \bG = \bS^{-1}\bW$ where $\bS$ communtes with $\bF$ and $\bW$ satisfies the orthogonal relations in the paper. We summarize the orthogonal relations as follow:

*   The zero blocks can be made orthonormal
*   The $(r, r-1)$ blocks are all orthogonal to the zero blocks.
*   The $(r, j)$ blocks are orthogonal to the orthogonal to the $(p, 0), (p-1, 0),\cdots, (r-j, 0)$ blocks.
The following code demonstrates the theorem.





```
from minimal_varx.minimal_varx import get_zero_blocks, LQ_multi_lags
from numpy.linalg import qr

def check_W(es, G):
    W, S = LQ_multi_lags(es, G)
    # print check decomposition:
    print("Check decomposition sum(abs(W - S @ G)) %f" % np.nansum(np.abs(W - S @ G)))
    FF = pm.calc_Jordan_matrix(es.Psi, 0)
    print("F:")
    print(FF)
    # print check commutative:
    print("check commutative sum(abs(FF @S - S @ FF)) %f" % np.nansum(np.abs(FF @ S - S @ FF)))
    W0 = get_zero_blocks(es, W)
    print("check orthogonal W0")
    print("sum abs W0 @ W0.T - I=%f" % np.nansum(np.abs(W0 @ W0.T - np.eye(es.agg_rnk))))
    print("check orthorgonal individuals:")
    n_psi = len(es.Psi)

    start_blk = 0
    for r in range(n_psi):
        rho, lx = es.Psi[r]
        for j in range(rho-1, -1, -1):
            if j == 0:
                start_blk += lx
                continue
            n_j_col = sum([rr[1] for rr in es.Psi if rr[0] >= rho - j])
            print(
                "rho=%d j=%d n_j_col=%d start=%d end=%d" % (
                    rho, j, n_j_col, start_blk, start_blk+lx))
            ee = np.nansum(
                np.abs(W[start_blk:start_blk+lx, :] @ W0[:n_j_col, :].T))
            if ee > 1e-5:
                print("BAD!!!")
            print(ee)
            start_blk += lx
    return W, S

Psi =  [(3, 2), (2, 1), (1, 2)]
d_0 = 1
m = sum([rr[1] for rr in Psi]) + d_0
es = varx_minimal_estimator(Psi, m=m)
np.random.seed(0)

G = np.random.randint(-10,10, (es.mm_degree, m))
G0 = get_zero_blocks(es, G)
# Check G0 is of full rank
print("full rank check %f" % np.prod(np.diagonal(qr(G0)[1])))
print("G")
print(G)
W, S = check_W(es, G)
print("W=%s" % str(W))
print("S=%s" % str(S))
```

Now we go through the examples. First is the case $\Psi = [(2, 1), (1, 1)]$ and $k=m=2$. In this case, we allocate one rank to exponent $2$ and one to expoment $1$, leaving $2-1-1=0$ rank free. We test for 10 simulations, the result in the table at the end shows the fitted results agree well with the original data. Here, we do not constraint $\bG$, just vectorize $\bG$ and use a standard optimizer.

```
# The following cell shows a number of simulations for the case where configuration of G is a half circle.
def k_2_p_2_sphere():
    np.random.seed(0)
    Psi = [(2, 1), (1, 1)]
    k_zero = 0
    k = sum([a[1] for a in Psi]) + k_zero
    p = Psi[0][0]
    T = 1000
    n = T + p
    n_test = 10
    results = np.zeros((n_test, 3))
    for ii in range(n_test):
        success = False
        cnt = 0
        while not success and cnt < 10:
          cnt += 1
          try:
            D = np.arange(k) * 0.2 + 1
            OO = random_orthogonal(k)
            es = varx_minimal_estimator(Psi, m=k)
            stable, H, F, G, Phi = gen_random_stable(es, k)
            Y, e = VAR_sim(Phi, n, D, OO)
            Y = Y.T
            e = e.T
            X = Y.copy()
            cov_res, cov_xlag, cov_y_xlag = calc_extended_covariance(Y, X, p)

            es_G = varx_minimal_estimator(Psi, m=k)
            es_G.set_covs(cov_res, cov_xlag)
            es_G.calc_states(G)
            if ~np.isnan(es_G.neg_log_llk):
              success = True
          except Exception:
            continue
          
        results[ii, 0] = es_G.neg_log_llk
        n_cnt = 0
        success = False        
        while (not success) and (n_cnt < 10):
          n_cnt += 1
          try:
            opt = es.hessian_fit(Y, X)
            if opt['success']:
                print('orig_llk %s est_llk=%s' % (
                    es_G.neg_log_llk, opt['fun']))                
                print(Phi.PolynomialMatrix_to_3darray())
                print(es.Phi)
                success = True
                results[ii, 1] = opt['fun']
                results[ii, 2] = 1                    
          except Exception as e:
            continue            
        if not success:
            print("failed for ii=%d" % ii)
            results[ii, 1] = opt['fun']
            results[ii, 2] = 0
    print(pd.DataFrame(results, columns=['org_llk', 'fitted_llk', 'success']))
                           
k_2_p_2_sphere()
```

```
Psi =  [(2, 1), (1, 1)]
d_0 = 0
m = sum([rr[1] for rr in Psi]) + d_0
es = varx_minimal_estimator(Psi, m=m)
np.random.seed(0)

G = np.random.randint(-10,10, (es.mm_degree, m))
G0 = get_zero_blocks(es, G)
# Check G0 is of full rank
print("full rank check %f" % np.prod(np.diagonal(qr(G0)[1])))
print("G")
print(G)
W, S = check_W(es, G)
print("W=%s" % str(W))
print("S=%s" % str(S))
```

As explained in the paper, if we constraint $\bG$ using the orthogonal relations, we are left with a sphere (with opposite points identified). By the orthogonalizing proposition, once we choose a vector of norm $1$ for $G_{1, 0}$, $G_{2, 0}$ must be orthogonal to it and is uniquely defined up to a scale of $\pm 1$, which does not change the result. Next, we graph the $llk$ versus $t$, where we represent the circle coordinates by $\cos t, \sin t$. Note that in this case
$$\bG = \begin{bmatrix} 0 & 0 \\ \cos t & \sin t \\ -\sin t & \cos t\end{bmatrix}$$
We see $\bW= \bS\bG$ in the above example is indeed of this form, obtained from $\bG$ without constraint.

```
# graphing llk vs the half circle
def plot_circle():
    np.random.seed(1)
    Psi = [(2, 1), (1, 1)]
    k_zero = 0
    k = sum([a[1] for a in Psi]) + k_zero
    p = Psi[0][0]
    T = 1000
    n = T + p
    D = np.arange(k) * 0.2 + 1
    OO = random_orthogonal(k)
    
    es_gen = varx_minimal_estimator(Psi, m=k)
    stable, H, F, G, Phi = gen_random_stable(es_gen, k)
    Y, e = VAR_sim(Phi, n, D, OO)
    Y = Y.T
    e = e.T
    X = Y.copy()
    cov_res, cov_xlag, cov_y_xlag = calc_extended_covariance(Y, X, p)
    
    es_G = varx_minimal_estimator(Psi, m=k)
    es_G.set_covs(cov_res, cov_xlag)
    es_G.calc_states(G)
    
    print('original llk %s' % es_G.neg_log_llk)    
    
    es_graph = varx_minimal_estimator(Psi, m=k)
    es_graph.set_covs(cov_res, cov_xlag)
    n_s = 100
    Theta = np.arange(n_s) * 1 * np.pi / n_s - np.pi / 2
    llks = np.zeros_like(Theta)
    
    def gen_gt(t):
        vv = np.array([np.cos(t), np.sin(t)])        
        return np.array([0, 0, vv[0], vv[1], -vv[1], vv[0]]).reshape(3, 2)

    for i in range(Theta.shape[0]):
        try:
            es_graph.calc_states(gen_gt(Theta[i]))
            llks[i] = es_graph.neg_log_llk
        except Exception as e:
            print(e)
            pass
    cosTheta = np.cos(Theta)
    # import pandas as pd
    # resDF = pd.DataFrame({'cosTheta': cosTheta, 'llk': llks}).sort_values('cosTheta')
    plt.plot(Theta, llks)
    plt.xlabel('Theta')
    plt.ylabel('-log_llk')
    plt.title('-log_llk vs Theta, org_llk=%f' % es_G.neg_log_llk)
    es_graph.simple_fit(Y, X)
    print(es_graph.Phi)
    print(Phi.PolynomialMatrix_to_3darray())
plot_circle()    
```

To illustrate the $LQ$ theorem in this case, we again show a general $G$ could be factorized to the $LQ$ form:

```
Psi =  [(2, 1)]
d_0 = 1
m = sum([rr[1] for rr in Psi]) + d_0
es = varx_minimal_estimator(Psi, m=m)
np.random.seed(0)

G = np.random.randint(-10,10, (es.mm_degree, m))
G0 = get_zero_blocks(es, G)
# Check G0 is of full rank
print("full rank check %f" % np.prod(np.diagonal(qr(G0)[1])))
print("G")
print(G)
W, S = check_W(es, G)
print("W=%s" % str(W))
print("S=%s" % str(S))
print("W @ W.T %s" % str(W @ W.T))
```

So $\bG$ could be orthogonalized. As explained in the paper this is the case of a circle and its tangent. In the example above we see the top row of $\bW$ is orthogonal to the second row.
The matrix $\bG$ is
$$\begin{bmatrix}-c\sin t & c\cos t \\ \cos t & \sin t

# The configuration of $\VAR(p)$ when m=k=2.
This could be consider the next step after from $AR(p)$. To fully study this case we probably need an article but we can use the LQ factorization to look at the search space for the orthogonalized $\bG^{o}$, which could be parametrized explicitly.

As $d_p = 2$ is the full rank case, we will assume $d_p = 1$. There are two main cases
*  $\Psi = [(p, 1), (\rho, 1)]$. Let $\bW = \bG^o$ be the orthogonalized $G$. For the first case, $\bW_{p, 0}$ and $\bW_{\rho, 0}$ form an orthogonal basis, $\bW_{\rho, \rho-1}\cdots \bW_{\rho, 1}$ are orthogonal to both vectors and hence is zero, $\bW_{p, p-1}\cdots \bW_{p, p-\rho}$ are also zero. The remaining $p-\rho-1$ vectors $\bW_{p, 1}\cdots \bW_{p, p-\rho-1}$ are proportional to $\bW_{\rho, 0}$. So this is a picture of the circle with $p-\rho-1$ tangent vectors. The minimal state space dimension is $p+\rho$. So the non reduced search space of $\bG$ has dimension $2(p+\rho)$, while the reduced one has dimension $2(p-\rho)$
*  $\Psi = [(p, 1)]$. In this case the $\bW_{p, p-1}\cdots \bW_{p, 1}$ are all orthogonal to $\bW_{p, 0}$. This is the case of a circle and $p-1$ tangent vectors. Here, the minimal state-space dimension is $p$, hen the non-reduced search space for $\bG$ is of dimension $2p$, while the reduced one has dimension $p$. 

We will not implement an optimizing routine here, but this case the search space could be parameterized explicitly as simple geometric object. The gradient on the reduced search space could be computed via the chain rule and we can use a standard optimzer to search. The next code segment illustrates the $LQ$ factorization discussed above.

```
from minimal_varx.minimal_varx import LQ_multi_lags
np.random.seed(1)
m = 2
p = 6
rho = 2
Psi = [(p, 1), (rho, 1)]
es = varx_minimal_estimator(Psi, m = 2)
G = np.random.randn(es.mm_degree, m)
W, S = LQ_multi_lags(es, G)
print("S=%s" % str(S))
print("W=%s" % str(W))
print("test factorization %s" % str(W - S @ G))
FF = pm.calc_Jordan_matrix(es.Psi, 0)
print("test commutative %s" % (FF @ S - S @ FF))
for j in range(p-1):
  print("ratio %d %s" % (j, str(W[j, :] / W[-1, :])))
```

```
def k_5_p_2_all_psi():
    np.random.seed(1)
    k = 5
    m = k
    p = 2
    Psi0 = [(2, 2), (1, 2)]

    T = 1000
    n = T + p
    D = np.arange(k) * 0.2 + 1
    OO = random_orthogonal(k)
    found_good_stable = False
    n_test = 10
    n_cnt = 0
    while (n_cnt < n_test) and (not found_good_stable):
        try:
            es = varx_minimal_estimator(Psi0, m=k)
            stable, H, F, G, Phi = gen_random_stable(es, k)
            Y, e = VAR_sim(Phi, n, D, OO)
            Y = Y.T
            e = e.T
            X = Y.copy()
            cov_res, cov_xlag, cov_y_xlag = calc_extended_covariance(Y, X, p)
            es_G = varx_minimal_estimator(Psi0, m=k)
            es_G.set_covs(cov_res, cov_xlag)
            es_G.calc_states(G)
            found_good_stable = True
        except Exception:
            pass
        n_cnt += 1
    
    all_psi = [a for a in list_all_psi_hat(m=m, p=p)]
    results = np.zeros((len(all_psi), p+3))
    header = [str(jj) for jj in range(p, 0, -1)]+['org_llk', 'fitted_llk', 'success']
    for ii in range(len(all_psi)):        
        Psi = psi_hat_to_psi(all_psi[ii])
        results[ii, :p] = all_psi[ii]
        es = varx_minimal_estimator(Psi, m=k)
        # print('X.shape=%s Y.shape=%s' % (str(X.shape), str(Y.shape)))
        # opt = es.gradient_fit(Y, X)
        print('Psi=%s' % Psi)
        results[ii, p] = es_G.neg_log_llk
        success = False
        n_cnt = 0
        while (not success) and (n_cnt < 10):
            opt = es.hessian_fit(Y, X)
            if opt['success']:
                print('orig_llk %s est_llk=%s' % (
                    es_G.neg_log_llk, opt['fun']))
                results[ii, p+2] = 1
                success = True
            results[ii, p+1] = opt['fun']
            n_cnt += 1
        if not opt['success']:
            print("failed opt=%s" % '\n'.join([jjj + ': ' + str(opt[jjj]) for jjj in opt.keys() if jjj not in ['hess_inv', 'jac', 'x']]))
            results[ii, p+1] = opt['fun']
    return pd.DataFrame(results, columns=header)
all_psi_table = k_5_p_2_all_psi()
print(all_psi_table)
print(all_psi_table.to_latex())
```

### Higher dimension and order
We test one with $k=8, p=3$. Convergence is slower, but this is because we optimize over the whole space instead of the reduced set satisfying the orthogonal relation.

```
from minimal_varx.utils import _calc_var_sim
def deautocorrel(e, p):
  k = e.shape[0]
  T = e.shape[1] - p
  eStack = np.zeros((e.shape[0]* p, T))      
  for i in range(p):
      eStack[k*i:k*(i+1), :] = e[:, p-i:p-i+T]
  Phi_e = np.linalg.solve(eStack @ eStack.T, eStack @ e[:, :T].T)
  e_ = e[:, 3:] - Phi_e[:8, :] @ e[:, :T]- Phi_e[8:16, :] @ e[:, 1:T+1] - Phi_e[16:, :] @ e[:, 2:T+2]
  return e_

def k_8_p3_generic():
    np.random.seed(2)
    Psi = [(3, 1), (2, 1), (1, 2)]
    k_zero = 4
    k = sum([a[1] for a in Psi]) + k_zero
    p = Psi[0][0]
    T = 2000
    n = T + p
    n_test = 10
    results = np.zeros((n_test, 4))

    for ii in range(n_test):
        n_tries = 10
        n_cnt = 0
        found_good_stable = False
        while (n_cnt < n_tries) and (not found_good_stable):
            try:
                D = np.arange(k) * 0.2 + 1
                OO = random_orthogonal(k)

                es = varx_minimal_estimator(Psi, m=k)
                stable, H, F, G, Phi = gen_random_stable(
                    es, k, scale_range=(0.1, .6))
                Y, e = VAR_sim(Phi, n, D, OO)
                # e_ = deautocorrel(e.T, p).T                        
                # Y = _calc_var_sim(Phi.PolynomialMatrix_to_3darray(), e_).T            
                Y = Y.T
                X = Y.copy()
                cov_res, cov_xlag, cov_y_xlag = calc_extended_covariance(Y, X, p)
                found_good_stable = True
            except Exception as ex:
                print(ex)
                pass
            n_cnt += 1
        print("found stable Phi for ii= %d" % ii)
        es_G = varx_minimal_estimator(Psi, m=k)
        es_G.set_covs(cov_res, cov_xlag)
        es_G.calc_states(G)
        results[ii, 0] = es_G.neg_log_llk
        results[ii, 2] = np.log(np.linalg.det(cov_res)) - np.log(np.linalg.det(cov_xlag))
        success = False
        n_cnt = 0
        while (not success) and (n_cnt < 11):
            opt = es.gradient_fit(Y, X)
            if opt['success']:
                print('orig_llk %s est_llk=%s' % (
                    es_G.neg_log_llk, opt['fun']))                
                print(Phi.PolynomialMatrix_to_3darray())
                print(es.Phi)
                success = True
                results[ii, 1] = opt['fun']
                results[ii, 3] = 1
            n_cnt += 1
        if not success:
            print("failed for ii=%d" % ii)
            results[ii, 1] = opt['fun']
            results[ii, 3] = 0
    return pd.DataFrame(results, columns=['org_llk', 'fitted_llk', 'full_llk', 'success'])
fitted_k8p3 = k_8_p3_generic()
```
